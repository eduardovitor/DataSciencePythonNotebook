Meu resumo sobre o que pode ser feito com a biblioteca Matplotlib

```python
import matplotlib.pyplot as plt
```

Gr√°fico de linha com Matplotlib


```python
x = [1,2]
y = [3,4]

plt.plot(x,y)
plt.show()
```

Colocando legenda no gr√°fico

```python
x = [1,2,5]
y = [2,3,7]

# T√≠tulo
plt.title("Meu primeiro gr√°fico com Python")

# Legendas nos eixos

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.plot(x,y)
plt.show()
```

Gr√°fico de barras

```python
x = [1,2,3,4,5]
y = [2,3,7,1,0]

titulo="Gr√°fico de barras"
eixox="Eixo X"
eixoy="Eixo Y"
# T√≠tulo
plt.title(titulo)

# Legendas nos eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x,y)
plt.show()
```

Comparando dois gr√°ficos de barra

```python
x1 = [1,3,5,7,9]
y1 = [2,3,7,1,0]

x2 = [2,4,6,8,10]
y2 = [5,1,3,7,4]

titulo="Gr√°fico de barras 2"
eixox="Eixo X"
eixoy="Eixo Y"
# T√≠tulo
plt.title(titulo)

# Legendas nos eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1,y1, label="Grupo 1")
plt.bar(x2,y2, label= "Grupo 2")
# Plotando legenda dos grupos
plt.legend()
plt.show()
```

Plotando o scatterplot

```python
x = [1,3,5,7,9]
y = [2,3,7,1,0]

titulo="Scatterplot"
eixox="Eixo X"
eixoy="Eixo Y"
# T√≠tulo
plt.title(titulo)
# Legendas nos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y)

plt.show()
```

Plotando scatterplot + gr√°fico de linhas

```python
x = [1,3,5,7,9]
y = [2,3,7,1,0]

titulo="Scatterplot"
eixox="Eixo X"
eixoy="Eixo Y"
# T√≠tulo
plt.title(titulo)
# Legendas nos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

Usando marcadores, cores e tamanhos diferentes nos gr√°ficos

```python
x = [1,3,5,7,9]
y = [2,3,7,1,0]

titulo="Scatterplot"
eixox="Eixo X"
eixoy="Eixo Y"
# T√≠tulo
plt.title(titulo)
# Legendas nos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)
 # color=cor ; marker=s√≠mbolo usado no gr√°fico (geralmente √© ponto) ; s=tamanho do marcador ; linestyle=estilo da linha
plt.scatter(x,y, label="Meus pontos", color="g", marker="h", s=100)
plt.plot(x,y, linestyle=":")
plt.legend()
plt.show()
```

Salvando figuras

```python
x = [1,3,5,7,9]
y = [2,3,7,1,0]

titulo="Scatterplot"
eixox="Eixo X"
eixoy="Eixo Y"
# T√≠tulo
plt.title(titulo)
# Legendas nos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)
 # color=cor ; marker=s√≠mbolo usado no gr√°fico (geralmente √© ponto) ; s=tamanho do marcador ; linestyle=estilo da linha
plt.scatter(x,y, label="Meus pontos", color="g", marker="h", s=100)
plt.plot(x,y, linestyle=":")
plt.legend()
plt.show()

plt.savefig("figura.pdf") # Salvando em pdf
plt.savefif("figura.png", dpi=300) # Salvando em png com qualidade boa (300 dpi)
```

Estudo de caso Crescimento da Popula√ß√£o da Brasileira

Lendo os dados sem libs

```python
dados = open("populacao_brasileira.csv").readlines()

x_anos=[]
y_pop=[]

for i in range(len(dados)):
  if i != 0:
    linha=dados[i].split(";")
    x_anos.append(int(linha[0]))
    y_pop.append(int(linha[1]))
```

Plotando os dados

```python

plt.plot(x_anos,y_pop)
plt.title("Crescimento da popula√ß√£o brasileira 1980-2016")
plt.xlabel("Anos")
plt.ylabel("Popula√ß√£o x 100 000 000")
plt.show()
```

Melhorando a visualiza√ß√£o

```python
plt.bar(x_anos,y_pop, color="grey")
plt.plot(x_anos,y_pop)
plt.title("Crescimento da popula√ß√£o brasileira 1980-2016")
plt.xlabel("Anos")
plt.ylabel("Popula√ß√£o x 100 000 000")
plt.show()
```

Criando boxplots
